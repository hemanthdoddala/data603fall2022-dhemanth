```python
import re
```

```python
def get_content_from_book(start, end, book, encoding):
    """Returns the content of the pages specified in the range.
    If start = 1 and end = 10 then this method returns the contents of pages in range 1 to 10 including 10
    this method expects 4 arguments 
    1. start - start page
    2. end - end page
    3. book- path of the book
    4. encoding - encoding required to open the file.
    """
    with open(book, 'r', encoding=encoding) as book4:
        book_content = book4.read();
    start_index = re.search(r'Page\ ?\|\ ?'+str(start), book_content).span()
    end_index = re.search(r'Page\ ?\|\ ?'+str(end+1), book_content).span()
    return book_content[start_index[0]:end_index[0]].strip()

def copy_content_to_file(start, end, book, encoding, dest_file):
    """This method extracts the data from the txt files and writes extracted content to new file.
    This method internally calls get_content_from_book method from,same class
    This method expects 5 arguments
    1. start - start page for extraction
    2. end - end page for extraction
    3. book - full path to book including file extension
    4. encoding - encoding required to open file 
    5. dest_file - name of the destination file, should include file extension as well 
        example:
        file1.txt
    """
    with open(dest_file, 'w+', encoding = 'utf-8') as file:
        file.write(get_content_from_book(start, end, book, 'utf-8'))
        print("Content is successfully copied to ", dest_file)
            
```

```python
%%file wordcount.py
# dob 08/08/1993

from mrjob.job import MRJob
import re

class WordCount(MRJob):
    
    def mapper(self, _, line):
        for word in line.split():
            # to remove unicode characters present in word
            word = word.encode("ascii", "ignore").decode()
            # to remove to non word characters like .,?_- and words are converted to lower case to exact occurrence
            word = re.sub(r"\W+|\_", '', word).lower().strip()
            if word != "":
                yield(word, 1)

    def reducer(self, word, counts):
         yield(word, sum(counts))
        
if __name__ == "__main__":
    WordCount.run()
```

```python
%%file countnonenglishwords.py

from mrjob.job import MRJob
import re
import enchant

class CountNonEnglishWords(MRJob):
    
    def mapper(self, _, line):
        d = enchant.Dict("en_US")
        for word in line.split():
            # to remove unicode characters present in word
            word = word.encode("ascii", "ignore").decode()
            # to remove to non word characters like .,?_- and words are converted to lower case to exact occurrence
            word = re.sub(r'([^\w\s]|_)+(?=\s|$)', '', word).lower().strip()
            word = re.sub(r"^\W+", "", word)
            if word != "" and d.check(word) != True:
                yield(word, 1)

    def reducer(self, word, counts):
         yield(word, sum(counts))
        
if __name__ == "__main__":
    CountNonEnglishWords.run()
```
